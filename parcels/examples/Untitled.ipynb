```python
from parcels import FieldSet
```

```python
data_path = 'NemoCurvilinear_data/'

filenames = {'U': data_path + 'U_purely_zonal-ORCA025_grid_U.nc4',
             'V': data_path + 'V_purely_zonal-ORCA025_grid_V.nc4',
             'mesh_mask': data_path + 'mesh_mask.nc4'}
variables = {'U': 'U', 'V': 'V'}
dimensions = {'lon': 'glamf', 'lat': 'gphif'}
field_set = FieldSet.from_nemo(filenames, variables, dimensions)
```

```python
import cartopy
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1, subplot_kw={'projection': cartopy.crs.Robinson()})

lons = field_set.U.grid.lon
lats = field_set.U.grid.lat
var = field_set.U.data
print lons.shape, lats.shape, var.shape

ax.pcolormesh(lons, lats, var[0, :, :], transform=cartopy.crs.PlateCarree())
plt.show()
```
